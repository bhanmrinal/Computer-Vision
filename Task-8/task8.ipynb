#### OBJECT DETECTION

```python
from ultralytics import YOLO
path='video.mp4'
model = YOLO('yolov8m.pt')
results = model(path, save=False)
frames = [i for i in range(1, 176)]
i=0

for r in results:
    print("Frame number ", frames[i])
    for box in r.boxes:
        class_id = r.names[box.cls[0].item()]
        print("Object type:", class_id)
    i+=1
```

#### COUNTING OBJECTS

```python
frames = [i for i in range(1, 176)]
i = 0
for r in results:
    frame_number = frames[i]
    object_counts = {'car': 0, 'person': 0}  
    for box in r.boxes:
        class_id = r.names[box.cls[0].item()]
        if class_id in object_counts:
            object_counts[class_id] += 1
    print("Frame number", frame_number)
    print("Number of cars:", object_counts['car'])
    print("Number of persons:", object_counts['person'])
    i += 1
```

#### COMPRESSING THE VID

```python
import cv2
import os
input_video_path = 'video.mp4'
output_video_path = 'cvideo.mp4'

video_capture = cv2.VideoCapture(input_video_path)
frame_width = int(video_capture.get(3))
frame_height = int(video_capture.get(4))
fps = int(video_capture.get(5))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = video_capture.read()
    if not ret:
        break
    out.write(frame)

video_capture.release()
out.release()

compressed_video_size_bytes = os.path.getsize(output_video_path)
video_size_MB = compressed_video_size_bytes / (1024 * 1024)
video_size_bytes = os.path.getsize(input_video_path)
compressed_video_size_MB = video_size_bytes / (1024 * 1024)  

print(f"Input Video Size: {video_size_MB:.2f} MB")
print(f"Compressed Video Size: {compressed_video_size_MB:.2f} MB")
```
