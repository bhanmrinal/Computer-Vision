##### Object Detection

```python
import matplotlib.pyplot as plt
```

```python
from ultralytics import YOLO
from PIL import Image

def obj_detection(results):
    result = results[0]
    object_info = []

    for box in result.boxes:
        class_id = result.names[box.cls[0].item()]
        coords = box.xyxy[0].tolist()
        coords = [round(x) for x in coords]
        confidence = round(box.conf[0].item(), 2)

        object_info.append({
            "Object type": class_id,
            "Coordinates": coords,
            "Probability": confidence
        })
    # Print object information
    for info in object_info:
        print("Object type:", info["Object type"])
        print("Coordinates:", info["Coordinates"])
        print("Probability:", info["Probability"])
        print("---")

    # Display the annotated image
    image = Image.fromarray(result.plot()[:, :, ::-1])
    plt.imshow(image)
    plt.show()

# Load a model
model = YOLO('yolov8n.pt')  # load an official model
```

```python
# Predict with the model
results = model("Input-01.webp")
obj_detection(results)
# Predict with the model
results = model("Input-02.webp")
obj_detection(results)
# Predict with the model
results = model("Input-03.jpg")
obj_detection(results)
```
